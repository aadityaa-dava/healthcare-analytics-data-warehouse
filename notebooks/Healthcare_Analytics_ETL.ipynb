## Missing Value Interpretation

Healthcare datasets frequently contain suppressed values to protect beneficiary privacy.

According to CMS reporting rules, metrics based on fewer than 11 beneficiaries may be suppressed and represented as null values.

Therefore, null values in this dataset do not always indicate data quality issues and must be interpreted within the healthcare reporting context.

## Preliminary Data Quality Decisions

| Category | Action |
|-----------|----------|
| Address Line 2 | Retain in raw layer, exclude from analytics |
| Suppression Indicators | Preserve as-is |
| Demographic Counts | Convert null values to 0 in curated layer |
| Chronic Condition Percentages | Preserve null values until feature engineering |
| Provider Name Fields | Retain |
| Financial Metrics | Retain |
| Geographic Fields | Retain |

## Fact Table Preparation

Before constructing the fact table, surrogate keys from the dimensions are mapped back to the source dataset.

This process establishes relationships between dimensions and the fact table and completes the star schema structure.

```python
location_lookup = dim_location[
    [
        "location_key",
        "city",
        "state_abbreviation",
        "state_fips",
        "zip_code",
        "ruca_code",
        "ruca_description",
        "country"
    ]
]
```

```python
claims_fact = claims.copy()
claims_fact = claims_fact.merge(
    location_lookup,
    left_on=[
        "Rndrng_Prvdr_City",
        "Rndrng_Prvdr_State_Abrvtn",
        "Rndrng_Prvdr_State_FIPS",
        "Rndrng_Prvdr_Zip5",
        "Rndrng_Prvdr_RUCA",
        "Rndrng_Prvdr_RUCA_Desc",
        "Rndrng_Prvdr_Cntry"
    ],
    right_on=[
        "city",
        "state_abbreviation",
        "state_fips",
        "zip_code",
        "ruca_code",
        "ruca_description",
        "country"
    ],
    how="left"
)
claims_fact.shape
```

```python
claims_fact["location_key"].isnull().sum()
```

## Fact Table Standardization

Prepare analytical measures for warehouse loading by handling suppressed and missing values according to CMS reporting conventions.

### Rules

- Count measures → replace null values with 0
- Financial measures → replace null values with 0
- Percentage metrics → preserve null values
- Risk metrics → preserve null values

```python
fact_provider_utilization.isnull() \
    .sum() \
    .sort_values(ascending=False) \
    .head(30)
```

## Warehouse Null Handling Strategy

The CMS dataset contains suppressed values used to protect beneficiary privacy.

### Count Measures

Null values are replaced with 0.

### Financial Measures

Null values are replaced with 0.

### Percentage Measures

Null values are preserved.

### Risk Measures

Null values are preserved.

### Suppression Indicators

Null values are preserved.

```python
count_measure_cols = [
    "Drug_Tot_HCPCS_Cds",
    "Drug_Tot_Benes",
    "Drug_Tot_Srvcs",
    "Med_Tot_HCPCS_Cds",
    "Med_Tot_Benes",
    "Med_Tot_Srvcs",
    "Bene_Age_LT_65_Cnt",
    "Bene_Age_65_74_Cnt",
    "Bene_Age_75_84_Cnt",
    "Bene_Age_GT_84_Cnt",
    "Bene_Feml_Cnt",
    "Bene_Male_Cnt",
    "Bene_Race_Wht_Cnt",
    "Bene_Race_Black_Cnt",
    "Bene_Race_API_Cnt",
    "Bene_Race_Hspnc_Cnt",
    "Bene_Race_NatInd_Cnt",
    "Bene_Race_Othr_Cnt",
    "Bene_Dual_Cnt",
    "Bene_Ndual_Cnt"
]
fact_provider_utilization[
    count_measure_cols
].isnull().sum().sum()
```

## Count Measure Standardization

Replace null values in additive count measures with zero.

This ensures accurate aggregation in SQL, PostgreSQL, and Power BI while preserving suppression behavior in percentage-based metrics.

```python
fact_provider_utilization[count_measure_cols] = (
    fact_provider_utilization[count_measure_cols]
    .fillna(0)
)
```

```python
fact_provider_utilization[
    count_measure_cols
].isnull().sum().sum()
```
